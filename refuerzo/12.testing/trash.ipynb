```python
fichero = open("coches.txt", "r", encoding="utf-8")
fichero_error = open("errores.txt", "w", encoding="utf-8")
linea_error = 1
_ = fichero.readline()
for linea in fichero:
    linea_error += 1
    linea = linea.strip()
    coche, cv = linea.split("#")
    try:
        cv = int(cv)
        if cv > 100:
            print(coche, cv)
    except:
        if len(cv) > 3:
            fichero_error.write(f"{linea_error}: {linea}\n")
        else:
            cv = cv[:-1]
        
fichero.close()
fichero_error.close()


```
