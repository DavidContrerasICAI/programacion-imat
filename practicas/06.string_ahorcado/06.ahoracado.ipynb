### Funcionalidad del programa

Se implementará el juego del Ahorcado.

Existirá una palabra clave de 5 letras a acertar, la cual se guardará en una variable. El juego proporcionará 5 opciones al usuario para acertar la palabra introducida indicando las pistas siguientes:
- Letra ROJA: la letra de esa posición no coincide con la de la palabra a acertar. 
- Letra VERDE: existe esa letra y está en la posición correcta.


El programa deberá comprobar que las letras son exactamente de 5 letras. En el caso de no ser así, se volverá a pedir (ver figura).

Además, se convertirán las palabras a letras mayúsculas (ver figura y descripción de la Iteración I).

### Implementación del programa

#### Incremento III - Colores

Se pintarán las letras de colores según el acierto de cada letra en su posición exacta. Ver la figura del juego. A continuación se muestra un ejemplo de cómo pintar cadenas de caracteres con colores.

````python
GRIS = "\033[1;47m"
VERDE = "\033[1;42m"
ROJO = "\033[1;41m"
DEFECTO = "\033[0m"

print(GRIS, "Inicio", VERDE,"Hola",ROJO, "Adiós", DEFECTO, "Me voy")
````

Intercalamos los colores mostrando solo espacios...

````python
print(VERDE,"S", DEFECTO, ROJO, "A", DEFECTO, VERDE,"L",DEFECTO, VERDE,"O",DEFECTO, ROJO, "N ")
````

Con strings formateados (f-string)...

````python
print(f"{VERDE} S {DEFECTO} {ROJO} A {DEFECTO} {VERDE} L {DEFECTO} {VERDE} O {DEFECTO} {ROJO} N ")
````
