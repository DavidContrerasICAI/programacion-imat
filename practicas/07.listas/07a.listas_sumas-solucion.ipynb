## Enunciado de la práctica

Desarrollar dos programas que realicen la siguiente funcionalidad.
### Práctica 7a1

Generar una lista aleatoria de 20 elementos en el rango [1, 10]. A continuación, se deberá obtener un número objetivo de forma aleatoria en el mismo rango [1, 10].

El programa seberá indicar las parejas de valores consecutivos de la lista que suman el número aleatorio objetivo.

La ejecución del programa deberá generar una salida similar a la siguiente. 

#### Salida
```
Lista: [8, 6, 1, 10, 8, 10, 4, 10, 5, 10, 2, 8, 4, 4, 5, 5, 6, 4, 6, 8]
Número objetivo: 10
2 + 8
5 + 5
6 + 4
4 + 6
```



## Solución 7a1

````python
import random

numeros = []
i = 0
while i < 20:
    numeros.append(random.randint(1, 10))
    i += 1

numero_objetivo = random.randint(1, 10)

print(f"Lista: {numeros}")
print(f"Número objetivo: {numero_objetivo}")
i = 0
while i < len(numeros)-1:
    if numeros[i] + numeros[i+1] == numero_objetivo:
        print(f"{numeros[i]} + {numeros[i+1]}")
    i += 1
````


### Práctica 7a2

Generar una lista aleatoria de 20 elementos en el rango [1, 10].
Obtener un número objetivo de forma aleatoria en el rango [1, 10].

Indicar qué parejas de valores **no repetidas** de la lista suman el número aleatorio objetivo. Veáse en la figura de ejemplo como se trata de no repetir las parejas de valores, por ejemplo la pareja "a" y la pareja "b" están formadas por los mismos números, 1 + 2, pero las parejas son distintas. No se debe mostrar la pareja "a" dos veces (repetida) de la forma 1 + 2 (en los índices -0 y 9-) y 2 + 1 (en los índices -9 y 0-).

La ejecución del programa deberá generar una salida similar a la siguiente. 


#### Salida
```
Lista: [1, 9, 5, 7, 9, 8, 6, 5, 3, 2, 6, 8, 1, 8, 2, 8, 4, 7, 4, 1]
Número objetivo: 3
1 + 2
1 + 2
2 + 1
2 + 1
1 + 2
2 + 1
```

## Solución 7a2

````python
import random

numeros = []

i = 0
while i < 20:
    numeros.append(random.randint(1, 10))
    i += 1

numero_objetivo = random.randint(1, 10)

print(f"Lista: {numeros}")
print(f"Número objetivo: {numero_objetivo}")

i = 0
while i < len(numeros)-1:
    j = i + 1
    while j < len(numeros):
        if numeros[i] + numeros[j] == numero_objetivo:
            print(f"{numeros[i]} + {numeros[j]}")
        j += 1
    i += 1
````
