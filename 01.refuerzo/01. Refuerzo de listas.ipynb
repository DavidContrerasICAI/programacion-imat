### Ejercicio #1. Intercambiar dos elementos de la lista (swapping)

Se pide intercambiar dos elementos de la lista. 
En el caso de poseer la siguiente lista:

["Luis", "Javier" , "Ana", "Teresa"]

Y se desea intercambiar la posición de "Luis" y "Ana", la lista resultante debería ser:

["Ana", "Javier", "Luis", "Teresa"]

Aunque Python permita realizar esta operación mediante una asignación doble como la que se muestre a continuación...
```python
personas = ["Luis", "Javier" , "Ana", "Teresa"]
persona1 = "Luis"
persona2 = "Ana"

pos1 = personas.index(persona1)
pos2 = personas.index(persona2)

personas[pos1], personas[pos2] = personas[pos2], personas[pos1]
print(personas)

['Ana', 'Javier', 'Luis', 'Teresa']
```
...se deberá programar esta solución mediante la algorítmica interna que emplearía Python por debajo para que el código sea extrapolable a cualquier lenguaje de programación.

El único método de list que se permite utilizar es index().

##### Pista #1: ¿Cómo crees que funciona internamente esa asignación?
Si se intenta hacer en dos pasos algo no funcionará...
```python
personas[pos1] = personas[pos2]
personas[pos2] = personas[pos1]
['Ana', 'Javier', 'Ana', 'Teresa']
```

¿Cómo lo resolverías?

````python
personas = ["Luis", "Javier" , "Ana", "Teresa"]
persona1 = "Luis"
persona2 = "Ana"

pos1 = personas.index(persona1)
pos2 = personas.index(persona2)

nombre_aux = personas[pos1]
personas[pos1] = personas[pos2]
personas[pos2] = nombre_aux

print(personas)
````

### Ejercicio #2. Implementar el funcionamiento del método remove(elemento)

Eliminar un elemento de una lista por su contenido de una forma manual, es decir, sin utilizar ninguno de los métodos que proporciona **list** ni operaciones de borrado (**del**).

**Input**: ["Luis", "Javier" , "Ana", "Teresa"] 

**Requisitos**: eliminar a "Javier"

**Output**: ["Luis" , "Ana", "Teresa"] 

Recordatorio: el método remove(elemento) borra el elemento de la lista cuyo contenido coincide con el elemento dado.

````python
personas = ["Luis", "Javier" , "Ana", "Teresa"]
print(personas)
persona_a_borrar = "Javier"
personas.remove(persona_a_borrar)
print(personas)
````

##### Pista #1: Primero pienso en cómo atacar este problema sin operar sobre la lista con funciones de borrado...
Dicho esto, podemos ver la posibilidad de crear una lista temporal y agregar todos los elementos excepto el buscado.

````python
personas = ["Luis", "Javier" , "Ana", "Teresa"]
lista_temporal = []
persona_a_borrar = "Javier"
i = 0
while i < len(personas):
    if personas[i] != persona_a_borrar:
        lista_temporal.append(personas[i])
    i += 1
    
personas = lista_temporal
print(personas)
````
