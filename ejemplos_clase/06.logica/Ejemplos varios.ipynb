```python
cadena = "camaleón"
i = 0
i_inv = -1
while i < len(cadena):
    print(f"{i}->{cadena[i]}", end = "")
    print(f"  {i_inv}->{cadena[i_inv]}")
    i += 1
    i_inv += -1
```
