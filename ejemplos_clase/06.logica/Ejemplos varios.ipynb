```python
cadena = "camaleón"
i = 0
i_inv = -1
while i < len(cadena):
    print(f"{i}->{cadena[i]}", end = "")
    print(f"  {i_inv}->{cadena[i_inv]}")
    i += 1
    i_inv += -1
```

### Ejemplo de diagrama de puertas lógicas

```python
A = 1
B = 0
C = 1
r1 = A ^ B    # XOR
S = r1 ^ C    # XOR
r2 = A & B    # AND
r3 = C & r1    # AND
C = r3 | r2    # OR
print(f"Para la entrada A={A} B={B} y C={C}")
print(f"La salida es S={S} y C={C}")
```
