### Estructura do-while

```python
palabra = input("Palabra: ")
while len(palabra) != 5:
    palabra = input("Palabra: ")
print(palabra)
```

### De estructura do-while a while

```python
palabra = ""    # Distinta a 5
while len(palabra) != 5:
    palabra = input("Palabra: ")
print(palabra)
```

### De un código de hace unos días...

```python
edad = 19
deporte = "futbol"
if edad >= 18:
    if deporte == "futbol":
        print("Eres mayor de edad y futbolista")
```

### ...a un código con operadores lógicos

```python
edad = 19
deporte = "futbol"
if edad >= 18 and deporte == "futbol":
    print("Te invito a una Coca-Cola")
```

### Negación de la condición

```python
edad = 15
deporte = "basket"
if not (edad >= 18 and deporte == "futbol"):
    print("Te invito a una Coca-Cola")
```

### Partimos de este código terriblemente malo

```python
edad = 19
deporte = "futbol"
if edad > 18:
    print("Te invito a una Coca-Cola")
if deporte == "futbol":
    print("Te invito a una Coca-Cola")
else:
    print("Te invito a una Coca-Cola")
```

### Código mejorado según ChatGPT

```python
edad = 19
deporte = "futbol"
# Usamos una sola condición con operadores lógicos
if edad > 18 or deporte == "futbol":
    print("Te invito a una Coca-Cola")
```
