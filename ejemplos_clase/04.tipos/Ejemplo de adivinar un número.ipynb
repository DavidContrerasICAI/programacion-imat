# Acierta un número

## Aproximación al problema

¿Sé pedir un número?

```python
numero_a_adivinar = 5    # En un futuro este número lo generaremos de forma aleatoria
numero_str = input("Dime un numero:")
```

¿Sé convertirlo?

```python
numero = int(numero_str)
```

¿Sé cuándo son iguales dos números?

```python
if numero == numero_a_adivinar:
    print("has acertado")
else:
    print("repite")
```

¿Sé hacerlo repetitivo/iterativo/..?

```python
while numero != numero_a_adivinar:
    .....
```

## Solución final

```python
numero_a_adivinar = 5    # En un futuro este número lo generaremos de forma aleatoria
numero_str = input("Dime un numero:")
numero = int(numero_str)
while numero != numero_a_adivinar:
    print("Has fallado")
    numero_str = input("Dime un numero:")
    numero = int(numero_str)

print("Has acertado") 
```

# Solución final mejorada
### Nunca hay que hacerlo perfecto a la primera

```python
numero_a_adivinar = 5    # En un futuro este número lo generaremos de forma aleatoria
numero = numero_a_adivinar + 1     # Fuerzo a un valor distinta (cualquiera) la primera vez
while numero != numero_a_adivinar:
    numero_str = input("Dime un numero:")
    numero = int(numero_str)
    print("Has fallado")

print("Has acertado") 
```
