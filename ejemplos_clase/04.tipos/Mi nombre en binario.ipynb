# Mi nombre en binario

```python
letra1 = ord("D")
letra2 = ord("A")
letra3 = ord("V")
letra4 = ord("I")
letra5 = ord("D")
```

```python
print(letra1, letra2, letra3, letra4, letra5)
```

```python
print(bin(letra1), bin(letra2), bin(letra3), bin(letra4), bin(letra5))
```

```python
print(hex(letra1), hex(letra2), hex(letra3), hex(letra4), hex(letra5))
```
