```python
class CodigoError(Exception):
    pass

def f1(producto:str) -> tuple[str, int]:
    nombre = producto[:3]
    codigo = producto[3:]
    try:
        codigo_int = int(codigo)
    except ValueError as error:
        raise CodigoError(codigo)
    return nombre, codigo_int

def f2() -> int:
    edad = "32a"
    try:
        edad_int = int(edad)
    except ValueError as error:
        edad_int = 11
    
    return edad_int

try:
    nombre, codigo = f1("NAR001")
    x = f2()
except CodigoError as error:
    print(error)
```
