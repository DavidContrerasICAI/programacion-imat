### Programa con funciones

```python
matriz = []

def crear():
    for _ in range(5):
        lista = []
        for _ in range(5):
            lista.append("____")
        matriz.append(lista)

def mostrar():
    for i in range(5):
        lista = []
        for j in range(5):
            print(matriz[i][j], end = " ")
        print()

def posicionar(objeto:str, fila:int, col:int):
    matriz[fila][col] = objeto 

def mover(objeto_a_mover:str, new_fila:int, new_col:int):
    for i in range(5):
        for j in range(5):
            if matriz[i][j] == objeto_a_mover:
                matriz[i][j] = "____"

    if matriz[new_fila][new_col] == "DANI":
        matriz[new_fila][new_col] = objeto_a_mover + "+D"
    else:
        matriz[new_fila][new_col] = objeto_a_mover
    
def buscar(objeto_a_mover:str) -> tuple[int, int]:
    coordenadas = (-1, -1)
    for i in range(5):
        for j in range(5):
            if matriz[i][j] == objeto_a_mover:
                coordenadas = i, j

    return coordenadas

def uber_to_persona(uber:str, pasajero:str):
    coordenadas_uber = buscar(uber)
    coordenadas_pasajero = buscar(pasajero)
    distancia_x = coordenadas_pasajero[0] - coordenadas_uber[0]
    distancia_y = coordenadas_pasajero[1] - coordenadas_uber[1]
    #distancia_total = distancia_x + distancia_y
    
    offset_x = 0
    offset_y = 0
    # Programado para que se avance paso a paso
    # Primero avanza X y luego Y
    if distancia_x > 0:   
        offset_x = +1
    elif distancia_x < 0:
        offset_x = -1
    else:
        if distancia_y > 0:
            offset_y = +1
        elif distancia_y < 0:
            offset_y = -1
            
    mover(uber, coordenadas_uber[0] + offset_x, coordenadas_uber[1] + offset_y)
```

```python
crear()
posicionar("UB1", 0, 0)
posicionar("DANI", 1, 3)
```

```python
mostrar()
```
