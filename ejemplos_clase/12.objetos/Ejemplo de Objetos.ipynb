```python
class Persona():
    def __init__(self, nombre:str, edad:int) -> None:
        if len(nombre) > 2:
            self.nombre = nombre.upper()
        else:
            raise ValueError("La longitud del nombre siempre >= 3")
            
        if edad >= 0:
            self.edad = edad
        else:
            raise ValueError("La edad siempre >= 0")
        
    def addEdad(self) -> int:
        self.edad += 1
    
    def isMayorEdad(self) -> bool:
        if self.edad >= 18:
            return True
        else:
            return False
    
    def __str__(self) -> str:
        return f"Nombre: {self.nombre} Edad: {self.edad}"
```
