```python
contador = 0
while contador < len(nombre):
    if nombre[contador] == " ":
        print("")
    else:
        if nombre[contador] == "a": 
            print("X", end = "")
        else:
            print(nombre[contador], end = "")
    contador = contador + 1    
```

```python
nombre = "Luis García Carrión"
nombre_modificado = ""
contador = 0
while contador < len(nombre):
    caracter = nombre[contador]
    if caracter == " ":
        nombre_modificado = nombre_modificado + "\n"
    else:
        if caracter == "a": 
            nombre_modificado = nombre_modificado + "X"
        else:
            nombre_modificado = nombre_modificado + caracter
    contador = contador + 1    
```

```python
print(nombre_modificado)
```

```python
indice = len(nombre) - 1
while indice >= 0:
    print(nombre[indice], end = "")
    indice = indice - 1
```
