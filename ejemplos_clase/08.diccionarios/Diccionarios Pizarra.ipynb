# Ejemplos de Diccionarios y estructuras avanzadas

## Trabajando con strings

```python
punto = "21,456"
x = int(punto[0:punto.index(",")])
y = int(punto[punto.index(",")+1:])
print(x, y)
```

```python
punto = "21,456"
x_str, y_str = punto.split(",")
x = int(x_str)
y = int(y_str)
print(x, y)
```

## Ejemplo completo de diccionarios

```python
alumnos = {2024001: ("Luis García", "C\\xxxx", {"Programación": 6, "Álgebra": 5})}
alumnos
```

```python
alumnos[2023001] = ("Ana Pascual", "C\\yyyy", {"Programación": 4, "Álgebra": 9})
```

```python
alumnos
```

```python
for clave, info in alumnos.items():
    nombre, direccion, notas = info
    print(nombre)
    print("="*20)
    for nombre_asig, nota_asig in notas.items():
        print(f" - {nombre_asig} {nota_asig}")
    print()
```

```python
_, _, notas = alumnos[2023001]
notas["Programación"] = 9
notas["Cálculo"] = 6
```

```python
for clave, info in alumnos.items():
    nombre, direccion, notas = info
    print(nombre)
    print("="*20)
    for nombre_asig, nota_asig in notas.items():
        print(f" - {nombre_asig} {nota_asig}")
    print()
```

```python
nombre, direccion, notas = alumnos[2023001]
direccion = "C\\Nueva dirección"
alumnos[2023001] = nombre, direccion, notas
```

```python
print(alumnos)
```
