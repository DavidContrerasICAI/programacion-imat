## Diccionarios

```python
coches = {}
type(coches)
```

```python
coches["Luis"] = "Seat"
coches["Ana"] = "Ferrari"
coches["Manolo"] = "BMW"
```

```python
coches["Ana"] = "dsdadhes"
```

```python
print(coches)
```

```python
coche_ana = coches["Ana"]
```

```python
coche_ana
```

```python
for personas in coches:
    print(personas)
```

```python
for personas in coches.keys():
    print(personas)
```

```python
for coches_list in coches.values():
    print(coches_list)
```

```python
coches.items()
```

```python
for tupla in coches.items():
    print(tupla[0])
```

```python
for tupla in coches.items():
    persona = tupla[0]
    coche = tupla[1]
    print(persona, "----->", coche)
```

```python
for tupla in coches.items():
    persona, coche = tupla
    print(persona, "----->", coche)
```

```python
for persona, coche in coches.items():
    print(persona, "----->", coche)
```

```python
if "ana" in coches:
    coches["ana"]
```
