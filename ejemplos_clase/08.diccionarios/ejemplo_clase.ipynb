## Práctica de polígonos

```python
poligonos = [ [(1,1), (2,2)], [(1,1), (2,2)], [(1,1), (2,2)] ]
poligonos_dict = { "pl1": [(1,2), (11,22)], "pl2": [(1,1), (2,2)], "pl3":[(1,1), (2,2)] }

lista_x = []
lista_y = []
for pol in poligonos_dict.values():
    for puntos in pol:
        x, y = puntos
        lista_x.append(x)
        lista_y.append(y)
        print(f"x:{x}, y:{y}")    
```
