# Juego de bombas

Implementa un juego de consola tipo buscaminas en un tablero cuadrado donde hay bombas y un premio oculto. El/la jugador/a introduce coordenadas hasta descubrir el premio (gana) o pisar una bomba (pierde). Al final, se muestra el tablero con colores ANSI.

### Objetivo
- Crear un tablero de tamaño 4 × 4 inicializado con espacios en blanco.
- Colocar aleatoriamente 3 bombas ("B") y un premio ("P") garantizando que el premio no coincida con una bomba.
- No es necesario evitar que dos bombas caigan en la misma casilla (puede haber sobrescritura).

Pedir coordenadas al usuario hasta que:
- Descubra el premio → WIN
- O caiga en una bomba → LOSE
- Si la casilla es vacía, márcala como visitada con CASI = "·" y sigue jugando.
- Al terminar, pinta el tablero usando fondo de color ANSI:
    + Bomba → ROJO
    + Premio → VERDE
    + Visitada (·) → AMARILLO
    + Vacía → GRIS

Resalta la última casilla jugada rodeando el símbolo con <>.

### Interacción y validación

Pide coordenadas X e Y con input() y valídalas hasta que estén en rango de la matriz.
Si la casilla elegida tiene una bomba o premio, termina la partida; si estaba vacía, márcala como "·" y continúa.
