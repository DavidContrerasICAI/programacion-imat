# Reto Bucles Carretera - Solución

Realizar una carretera como la que se muestra en la salida que vaya apareciendo poco a poco. Para conseguir este efecto (como se aprecia en el vídeo), se trabajará con la librería time.

En el siguiente ejemplo, se introducirá un retardo de 0.1 segundos entre el texto "Hola" y "Adiós".

```
import time
print("Hola")
time.sleep(0.1)
print("Adiós")
```


````python
import time

offset = 1
curva = 0
i = 0
while i < 30:
    j = 0 
    while j < curva:
        print(" ", end="")
        j += 1
    print("|", end="")
    j = 0
    while j < 5:
        print("#", end="")
        j += 1
    print("|", end="")
    curva += offset
    if curva == 5 or curva == 0:
        offset *= -1

    print()
    i += 1
    
    time.sleep(0.1)
````
